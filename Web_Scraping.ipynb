## 웹 스크래핑과 웹 크롤링

**웹 스크래핑**
- 웹 사이트 상에서 원하는 정보를 추출하는 기술

**웹 크롤링**
- 웹 크롤러(자동화 봇)가 일정 규칙으로 웹 페이지를 브라우징하는 것

**주의점**
- 서버에 과부하를 주는 등의 피해를 줄 수 있으므로, 법적인 문제에 대해 전문적으로 알아볼 것
- 특히, 학습용이 아닌 상용을 목적으로 크롤링 및 스크래핑시에는 더욱 주의할 것

## 웹 페이지란?

HTML 로 만들어진 문서

### HTML 이란?

**Hyper Text Markup Language**
- 텍스트를 뛰어 넘는 표시 언어
- 텍스트 정보에 대한 순차적 접근을 뛰어넘음
- 링크 등을 통해 클릭 한 번으로 다양한 페이지로 한번에 이동 가능
- 마크업 언어를 통해 문서를 구조적으로 표현 가능

### HTML 요소(Elements)

- 태그(Tag)
- 값(Value)
- 속성(Attribute)

**태그(Tag)**
- 문서의 구조를 표현하는 구성 요소
- 꼬리표(Tag)를 이용하여 문서의 구조를 표현함
- 짝꿍이 있는 태그
    - `<a href="https://google.com">Google</a>`
    - `<div id="container"></div>`
    - `<span class="word"></span>`
- 나홀로 태그
    - `<img src="c:/abc.jpg" />`
    - `<input type="text" />`


**값(Value)**
- 태그의 명령을 받는 데이터
- 데이터 그 자체
- 대부분 여는 태그와 닫는 태그 사이에 존재

**속성(Attribute)**
- 태그를 보조하는 구성 요소
- 태그 안쪽에 기입
- 속성명과 속성값으로 구성됨

```
<p style="color:red;">빨간색 글씨</p>
<p style="font-size:20px;">폰트 사이즈</p>
```

## 정적 웹 페이지와 동적 웹 페이지

**정적 웹 페이지**
- 서버에 미리 저장된 파일(HTML, 이미지, JavaScript 파일 등)이 그대로 전달되는 웹 페이지
- 사용자는 고정된 웹 페이지를 보게 됨

**동적 웹 페이지**
- 서버에 있는 데이터가 JavaScript에 의해 가공/처리된 후 전달되는 웹 페이지
- 사용자는 특정한 상황, 시간, 요청 등에 따라 달라지는 웹 페이지를 보게 됨

## 자바스크립트란?

**JavaScript**
- 웹 페이지와 상호작용하도록 만들어진 스크립트 언어
- HTML 문서 내에 `<script></script>` 태그를 통해 삽입
- 웹 브라우저를 통해 동작됨
